# Matplotlib 

1) Matplotlib is a popular Python library for making visualizations (charts, graphs, and plots).
2) It’s especially good for 2D plots like line graphs, bar charts, scatter plots, histograms, pie charts, etc.

```python
import matplotlib.pyplot as plot
```

# Line Graph

1) A line graph is used to show trends over time or continuous data.
2) It connects data points with a line, making it easy to see increases, decreases, or patterns.

```python
x=[10,20,30,40,50,60]
y=[0,200,400,600,800,1000]

plot.figure(figsize=(3,3))
plot.plot(x,y,color='red',marker='o',ls='--')
plot.xlabel('x-axis')
plot.ylabel('y-axis')
plot.title("Line Graph")
plot.show()
```

# Scatter Plot

1) Shows relationships between two variables.
2) Example: Height vs. Weight, Study Hours vs. Scores.
3) Reveals patterns, trends, or correlations.
4) Positive correlation → points slope upward.
5) Negative correlation → points slope downward.
6) No correlation → points scattered randomly.
7) Highlights clusters and outliers in data.
8) Points are not connected by lines

```python
x=[32,65,76,87,32,54]
y=[132,434,323,554,654,435]

plot.figure(figsize=(3,3))
plot.scatter(x,y,color='black',marker='*')
plot.xlabel('x-axis')
plot.ylabel('y-axis')
plot.title("Scatter Graph")
plot.show()
```

# pie graph

1) A pie chart is best for showing proportions in a dataset as slices of a circle.

```python
x=[100,200,300,400,500]
y=['ab','cd','ef','gh','ij']

plot.figure(figsize=(3,3))
exp=[0,0,0,0,0,]
plot.pie(x,labels=y,explode=exp,shadow=True,autopct='%1.1f%%')
plot.title('Pie Graph')
plot.legend(title='ABC',loc='right')
plot.show()
```

# Bar graph

1) A bar graph (or bar chart) is a chart that uses rectangular bars to represent data.
2) The length/height of each bar shows the value of each category.
3) Bars can be vertical or horizontal.

```python
# verticle bar graph
x=[23,43,34,43,54,54]
y=['ac','as','df','sd','gh','ld']
plot.title('Verticle Bar graph')
plot.xlabel("sale")
plot.ylabel("categories")
plot.bar(y,x,width=0.4,color='pink')
plot.show()
```

```python
# horizontal bar graph
x=[10,20,25,30,32]
y=['sd','dp','ds','da','hj']

plot.title("horizontal bar graph")
plot.xlabel("x-axis")
plot.ylabel("y-axis")
plot.barh(y,x,height=0.5,color='red')
plot.show()
```

# Histrogram 

1) A histogram is a graph that shows the distribution of data.
2) It looks similar to a bar graph but instead of categories, the x-axis is divided into intervals (bins).

```python
x=[45, 56, 60, 62, 65, 68, 70, 71, 72, 74, 75, 77, 80, 82, 83, 85, 86, 88, 90, 95]

plot.title('Histrogram')
plot.hist(x,bins=8,edgecolor='black')
plot.show()
```

# Subplot -
**A subplot is a feature in Matplotlib that allows you to place multiple plots within a single figure.**

```python
x=[20,30,43,53,45]
y=[54,67,87,98,89]

plot.figure(figsize=(10,3))
plot.subplot(1,2,1)
plot.title("Line Graph")
plot.xlabel("x-axis")
plot.ylabel("y-axis")
plot.plot(x,y,marker='o')

plot.subplot(1,2,2)
plot.title("Scatter Graph")
plot.xlabel("x-axis")
plot.ylabel("y-axis")
plot.scatter(x,y)
plot.show()
```

**example- that creates a figure with a 1x3 grid of subplots: a line graph, a scatter plot, and a bar chart**

```python
plot.figure(figsize=(10,3))

plot.subplot(1,3,1)
x=[100,200,300,400,500]
y=[10,20,30,40,50]
plot.title("Line graph")
plot.xlabel("x-axis")
plot.ylabel("y-axis")
plot.plot(x,y,marker='o')

plot.subplot(1,3,2)
x=[34,54,76,87,54]
y=[43,76,85,43,32]
plot.title("Scatter graph")
plot.scatter(x,y,marker='*')

plot.subplot(1,3,3)
x=[34,55,54,33,43]
y=['fcg','rgd','gtf','tgd','tgd']
plot.title("Bar graph")
plot.bar(y,x,width=0.5)

plot.show()
```
