```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

**Create dataframe**

```python
df=sns.load_dataset("tips")
df
```

# Line graph

```python
sns.lineplot(x='total_bill',y='tip',data=df,hue='sex')
```

# Scatter graph

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',sizes=(5,5))
```

# Bar Graph

```python
sns.barplot(x='day',y='total_bill',data=df,hue='sex',errorbar=None)
```

# Histrogram

```python
sns.histplot(x='total_bill',data=df,bins=5,hue='sex')
```

# Density graph (kde plot)

```python
sns.kdeplot(x='total_bill',data=df,hue='sex')
```

# Boxplot

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='size',y='tip',data=df,hue='day')
```

# Violin plot

```python
sns.violinplot(y='total_bill',x='day',data=df,hue='day')
```

# Heatmap graph

```python
co=df.corr(numeric_only=True)
sns.heatmap(co,annot=True)
```

# Pair plot

```python
sns.pairplot(data=df,hue='day')
```

# Count Plot

```python
sns.countplot(x='day',data=df,hue='time')
```

```python
import numpy as np

data = np.array([5, 5, 6, 7, 8, 9, 12, 14, 15, 18])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers:", outliers)
```
