```python
np.concatenate((j1,j2))  # joinning 1-d array 
```

```python
np.concatenate((j3,j4))       # joinning 2-d array  
                               # syntax for - concatenate((aaray1,array2),axis)
                               # axis-0 = column wise joinning (Defalut axis-0)
                               # axis-1 = row wise joinning
```

```python
np.concatenate((j4,j3),axis=1)
```

```python
import numpy as np
import time
list1=list(range(1,10000001))
list2=list(range(1,10000001))
start=time.time()
result=[list1[i]+list2[i] for i in range(len(list1))]
end=time.time()
list_time=end-start



array1=np.arange(1,10000001)
array2=np.arange(1,10000001)
start1=time.time()
result=array1+array2
end1=time.time()
array_time=end1-start1


if array_time<list_time:
    print("array is faster")
else:
    print("list is faster")
```
